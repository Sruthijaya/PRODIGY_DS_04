OBJECTIVE

Aim to analyze sentiment patterns in social media data related to the Borderlands franchise By categorizing tweets into positive, negative, and neutral sentiments

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV file
df = pd.read_csv('twitter_training.csv')

# Convert 'Sentiment' column to categorical data type
df['SENTIMENT'] = df['SENTIMENT'].astype('category')

# Count sentiments
sentiment_counts = df['SENTIMENT'].value_counts()

# Plot
plt.figure(figsize=(8, 6))
sentiment_counts.plot(kind='bar', color=['green', 'blue', 'red'])
plt.title('Sentiment Analysis of Borderlands Tweets')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

CONCLUSION


This shows the sentiment distribution of tweets about Borderlands. The x-axis shows the sentiment, which can be positive, neutral, irrelevant, or negative. The y-axis shows the count of tweets.
- Positive sentiment tweets (green) have the highest count, followed by neutral tweets (blue).
- There are fewer irrelevant tweets (red) and negative sentiment tweets (grey).
- This suggests that the overall sentiment of the tweets about Borderlands is positive.
